```python
pd.pivot_table(df,index=['sex','pclass'],values='age',aggfunc='mean') #missing values ka isse pehle q hain karlena
```

```python
df['t_age']=round(df.groupby(['sex','pclass'])['age'].transform('mean'),0)
```

```python
df["final_age"] = np.where((df.age.isna()), df.t_age, df.age)
df
```

```python
df.drop(columns = ["t_age", "age"], inplace = True )
```
